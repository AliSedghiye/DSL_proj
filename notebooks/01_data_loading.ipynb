```python
import pandas as pd
import numpy as np
import os

# Confirm you can see the file
print(os.listdir("../data/raw"))
```

```python
df = pd.read_csv("../data/raw/Database_ENG.csv")
print(f"âś… Loaded: {df.shape[0]} rows, {df.shape[1]} columns")
df.head(3)
```

```python
print("=== COLUMN NAMES ===")
print(df.columns.tolist())

print("\n=== DATA TYPES ===")
print(df.dtypes.value_counts())

print("\n=== MISSING VALUES (top 20) ===")
print(df.isnull().sum().sort_values(ascending=False).head(20))
```

```python
# In IACOFI: -97=don't know, -98=not applicable, -99=refused, -999=irrelevant answer
MISSING_CODES = [-97, -98, -99, -999]
df.replace(MISSING_CODES, np.nan, inplace=True)

print("âś… Negative codes replaced with NaN")
print(f"Missing values per key variable:")
key_vars = ['qd1','qd2','qd7','qd9','qd13','qd14','qd10']
print(df[key_vars].isnull().sum())
```

```python
# ISTAT 2023 official statistics â€” merged at macro-region level
regional = pd.DataFrame({
    'qd2':                 [1,            2,            3,        4,       5       ],
    'region_name':         ['North-West', 'North-East', 'Center', 'South', 'Islands'],
    'unemployment_rate':   [4.7,          3.8,          6.5,      14.0,    13.5    ],
    'internet_penetration':[86.3,         86.1,         85.0,     78.0,    79.5    ]
})
df = df.merge(regional, on='qd2', how='left')
print(f"âś… Regional context merged")
print(df[['region_name','unemployment_rate','internet_penetration']].value_counts('region_name'))
```

```python
df.to_csv("../data/processed/iacofi_clean.csv", index=False)
print("âś… Saved to data/processed/iacofi_clean.csv")
print(f"Final shape: {df.shape}")
```
